```python
import pandas as pd
import requests
pd.set_option('display.max_columns', None) # so we can see all columns in a wide dataframe
import numpy as np
```

```python
url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2003-04&SeasonType=Regular%20Season&StatCategory=PTS'

r = requests.get(url).json()
r
```

```python
column_headers = r['resultSet']['headers']
column_headers
```

```python
rows = pd.DataFrame(r['resultSet']['rowSet'])
df1 = pd.DataFrame(r['resultSet']['rowSet'], columns = column_headers)
df1
```

```python
df2 = pd.DataFrame({'year':['2003-2004' for i in range (len(df1))],
                    'Season_type':['Regular%20Season' for i in range (len(df1))]})
df3 = pd.concat([df2,df1], axis=1)
df3
    
```

```python
# Delete the dataframes made so we can start our for loop
del df1, df2, df3
```

```python
# for loop to concat all the df3's for each year starting from 2003-04 until 2022-23

base_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season={season}&SeasonType=Regular%20Season&StatCategory=PTS'
for i in range(20):
    #calculate the starting year and ending year of the season
    start_year = 2003 + i
    end_year = 2004 + i
    season_str =f'{str(start_year)}-{str(end_year)[-2:]}'
    url = base_url.format(season=season_str)
    r = requests.get(url).json()
    df1 = pd.DataFrame(r['resultSet']['rowSet'], columns = column_headers)
    df2 = pd.DataFrame({'year':[season_str for i in range (len(df1))],
                    'Season_type':['Regular%20Season' for i in range (len(df1))]})
    if i == 0:
         df3 = pd.concat([df2,df1], axis=1)
    else:
        df3 = pd.concat([df3,pd.concat([df2,df1], axis=1)], axis=0).reset_index(drop=True)
df3
```

```python
# convert ddf3 to csv file
df3.to_csv("NBAstats_2003-2023_regularseason.csv")
```
